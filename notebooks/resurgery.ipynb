```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("data/out/re_surgery.csv")
df.head(5)
```

```python
plt.figure(figsize=(10, 6))

# Count how many patients had repeats for each surgery type
order = df['procedure_name'].value_counts().iloc[:10].index # Top 10 only

sns.countplot(
    data=df, 
    y='procedure_name', 
    order=order,
    palette='viridis'
)

plt.title('Top 10 Procedures Requiring Re-Surgery (By Patient Count)')
plt.xlabel('Number of Patients affected')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.countplot(
    data=df, 
    x='occurrence_count', 
    palette='magma'
)

plt.title('Distribution of Surgery Counts')
plt.xlabel('Number of Surgeries Performed on Same Patient')
plt.ylabel('Count of Patients')
plt.show()
```
